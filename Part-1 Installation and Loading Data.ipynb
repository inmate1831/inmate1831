### 6. Viewing Certain Rows Only

#### 6.1 .head() Method

```python
# head() method can be used to return first n rows
# usually, it returns first five rows
df.head()
# however, a value can be passed into parenthesis
# for e.g., df.head(10) will return first 10 rows
```

#### 6.2 .tail() Method

```python
# tail() method can be used to return last n rows
# usually, it returns last five rows
df.tail()
# however, a value can be passed into parenthesis
# for e.g., df.tail(10) will return last 10 rows
```
